##NLP

removing stop words, puncuation mark and lemetization

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words=set(stopwords.words('english'))
lemmatizer=WordNetLemmatizer()
```

```python
def pre_process_text(text):
  text=str(text).lower() #lowercase
  text=re.sub(r'[^\w\s]','',text) #removing punchuation
  words=text.split()
  words=[lemmatizer.lemmatize(word) for word in words if word not in stop_words] #lemetization and skipping stop words
  return " ".join(words)
```

```python
df['tags']=df['tags'].apply(pre_process_text)
```

```python
df['tags'][0]
```

vectorization

```python
df=df.reset_index(drop=True) #so that there are no missing index after removing duplicates
```

```python
indices=pd.Series(df.index,index=df['title'])
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(max_features=50000,stop_words='english',ngram_range=(1,2))
tfidf_matrix=tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix
```

Using Cosine Simmilerity to find movies

<img src="https://wikimedia.org/api/rest_v1/media/math/render/svg/15d11df2d48da4787ee86a4b8c14551fbf0bc96a">

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recomend(title, n=10):
  if title not in indices:
    return ['Movie not found']

  idx = indices[title]
  sim_score=cosine_similarity(tfidf_matrix[idx],tfidf_matrix).flatten()
  sim_idx = sim_score.argsort()[::-1][1:n+1]

  return df['title'].iloc[sim_idx]
```

```python
recomend('Avatar')
```

```python
recomend('spider-man')
```
